```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the image path
image_path = '/content/Box_Filter.jpg'

# Check if the file exists
if not os.path.isfile(image_path):
    raise ValueError(f"Image not found at {image_path}. Please check the file path.")

# Load the image
i = cv2.imread(image_path)
if i is None:
    raise ValueError("Image not found or unable to load.")

# Create a Box Filter Kernel (7x7)
kernel_size = 7
w = np.ones((kernel_size, kernel_size)) / (kernel_size * kernel_size)

# Apply Box Filter
gd = cv2.filter2D(i, -1, w)

# Display Original and Box Filtered Images
plt.figure(figsize=(15, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))
plt.title('Original Picture')
plt.axis('off')  # Hide axes

# Box Filtered Image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gd, cv2.COLOR_BGR2RGB))
plt.title('Box Filtered Image')
plt.axis('off')  # Hide axes

# Show the plots
plt.show()
```
