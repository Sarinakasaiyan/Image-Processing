```python
#sarinakasaiyan

from skimage.io import imread
from skimage import exposure
import matplotlib.pyplot as plt

# Load the image
i = imread('/content/Adjust_Intensity_Values.jpg')
# Adjust intensity values
j = exposure.rescale_intensity(i, in_range=(0, 125), out_range=(0, 1))

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Display the original image
axs[0].imshow(i)
axs[0].set_xlabel('Original', fontsize=12)  # Set title at the bottom

# Display the adjusted image
axs[1].imshow(j)
axs[1].set_xlabel('Adjusted', fontsize=12)  # Set title at the bottom

# Remove axis ticks
plt.axis('off')

# Show the plot
plt.show()
```
