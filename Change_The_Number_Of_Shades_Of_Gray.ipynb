```python
#sarinakasaiyan
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the RGB image
RGB = cv2.imread('/content/Change_The_Number_Of_Shades_Of_Gray.jpg')

# 2. Convert the RGB image to grayscale
I = cv2.cvtColor(RGB, cv2.COLOR_BGR2GRAY)

# 3. Normalize the grayscale image to 128 levels
I_128 = np.uint8(I * (128 / 256))

# 4. Normalize the grayscale image to 64 levels
I_64 = np.uint8(I * (64 / 256))

# 5. Display the results
plt.figure(figsize=(15, 8))

# Display the original RGB image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(RGB, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis("off")

# Display the grayscale image with 128 levels
plt.subplot(1, 3, 2)
plt.imshow(I_128, cmap='gray')
plt.title('Grayscale (128 levels)')
plt.axis("off")

# Display the grayscale image with 64 levels
plt.subplot(1, 3, 3)
plt.imshow(I_64, cmap='gray')
plt.title('Grayscale (64 levels)')
plt.axis("off")

# Show all images
plt.show()
```
