```python
#sarinakasaiyan

import cv2
import numpy as np
# Import the necessary function for displaying images in Colab
from google.colab.patches import cv2_imshow

# Read the image
img = cv2.imread('/content/Gaussian-Filter.png')

# Create gaussian filter
filter_size = (5, 5)
sigma = 3
gaussian_filter = cv2.getGaussianKernel(filter_size[0], sigma) * cv2.getGaussianKernel(filter_size[1], sigma).T

# Applying the Gaussian filter to the Image
filtered_img = cv2.filter2D(img, -1, gaussian_filter)

# Display the original image and the reconstructed image
# Use cv2_imshow instead of cv2.imshow
cv2_imshow(img)
cv2_imshow(filtered_img)
# cv2.waitKey(0) # This is not needed in Colab
# cv2.destroyAllWindows() # This is not needed in Colab
```
