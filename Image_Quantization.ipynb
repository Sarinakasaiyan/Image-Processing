```python
#sarinakasaiyan
from PIL import Image
import matplotlib.pyplot as plt

# Function to save and display images with different qualities
def save_and_display_image(image_path, qualities):
    # Read the image
    original_image = Image.open(image_path)

    # Create a list to store filenames
    saved_files = []

    # Quantize the image with different qualities
    for quality in qualities:
        file_name = f'Quality_{quality}.jpg'
        original_image.save(file_name, quality=quality)
        saved_files.append(file_name)

    # Display images
    fig, axs = plt.subplots(1, len(saved_files), figsize=(15, 5))
    for ax, file in zip(axs, saved_files):
        img = plt.imread(file)
        ax.imshow(img)
        ax.axis('off')  # Hide axes for a cleaner look
        ax.set_title(f'Quality {file.split("_")[1].split(".")[0]}')

    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()

# Specify image path and desired qualities
image_path = "/content/Image_Quantization.jpg"
qualities = [40, 16, 5]

# Call the function
save_and_display_image(image_path, qualities)
```
