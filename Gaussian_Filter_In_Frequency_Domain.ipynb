```python
#sarinakasaiyan
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import fftpack

# Read the image
img = cv2.imread('/content/Gaussian_Filter_In_Frequency_Domain.png', 0)

# Fourier transform
fft_image = fftpack.fft2(img)

# Shifted Fourier transform
shifted_fft_image = fftpack.fftshift(fft_image)

# Return to the initial state (inverse shift)
filtered_shifted_fft_image = fftpack.ifftshift(shifted_fft_image)

# Image reconstruction using inverse Fourier transform
reconstructed_image = fftpack.ifft2(filtered_shifted_fft_image)

# Display the original image and the reconstructed image
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Picture')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(np.log(np.abs(fft_image) + 1), cmap='gray')
plt.title('Fourier Transform (Magnitude)')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(np.log(np.abs(shifted_fft_image) + 1), cmap='gray')
plt.title('Shifted Fourier Transform (Magnitude)')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(np.log(np.abs(filtered_shifted_fft_image) + 1), cmap='gray')
plt.title('Inverse Shifted Fourier Transform (Magnitude)')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(np.abs(reconstructed_image), cmap='gray')
plt.title('Restored Picture')
plt.axis('off')

plt.show()
```
