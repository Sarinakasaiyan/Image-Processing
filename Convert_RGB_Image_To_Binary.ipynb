```python
#sarinakasaiyan
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the RGB image
I = cv2.imread('/content/Convert_RGB_Image_To_Binary.ipynb.jpg')

# Convert the image from BGR to RGB for display
I_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

# Convert the image from BGR to grayscale
J = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

# Apply thresholding to convert grayscale to binary image
threshold = 0.7  # Adjust this threshold value as needed
_, J = cv2.threshold(J, int(threshold*255), 255, cv2.THRESH_BINARY)

# Display the input RGB image and output binary image
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.imshow(I_rgb)
ax1.set_title('Input RGB Image')
ax1.axis('off')

ax2.imshow(J, cmap='gray')
ax2.set_title('Output Binary Image')
ax2.axis('off')

plt.tight_layout()
plt.show()
```
