```python
#sarinakasaiyan
import cv2
import matplotlib.pyplot as plt

# Read the image
i = cv2.imread("/content/Cropped.jpg")

# Crop the region of interest (change the coordinates for a different crop)
cr = i[100:450, 150:450]  # Changed crop coordinates

# Apply Gaussian Blur to the cropped image
cr_blurred = cv2.GaussianBlur(cr, (5, 5), 0)

# Save the cropped and blurred image
cv2.imwrite("/content/Cropped_Blurred.jpg", cr_blurred)

# Display the original and cropped images
plt.figure(figsize=(10, 5))  # Set figure size for better visibility

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))  # Corrected color conversion for original
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(cr, cv2.COLOR_BGR2RGB))
plt.title('Cropped')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(cr_blurred, cv2.COLOR_BGR2RGB))
plt.title('Cropped & Blurred')
plt.axis('off')

plt.show()
```
