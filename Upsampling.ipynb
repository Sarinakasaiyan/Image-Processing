```python
#sarinakasaiyan
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('/content/Upsampling.jpg')

# Upsample the image by a factor of 2 using bilinear interpolation
upsampled_img = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

# Save the upsampled image
cv2.imwrite('/content/Upsampled_Image.jpg', upsampled_img)

# Display the original and upsampled images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis("on")  # Keep axes on

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(upsampled_img, cv2.COLOR_BGR2RGB))
plt.title('Upsampled Image')
plt.axis("on")  # Keep axes on

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
