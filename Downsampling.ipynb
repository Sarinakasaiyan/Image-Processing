```python
#sarinakasaiyan
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Import cv2_imshow for displaying images in Colab

# Read the image
image = cv2.imread('/content/Downsampling.jpg')

# Check if the image was loaded successfully
if image is None:
    raise ValueError("Failed to load image. Check if the file is valid and accessible.")

# Downsample by a factor of 1/2
downsampled_image_half = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)

# Downsample by a factor of 1/4
downsampled_image_quarter = cv2.resize(image, None, fx=0.25, fy=0.25, interpolation=cv2.INTER_NEAREST)

# Display the original and downsampled images using cv2_imshow
cv2_imshow(image)  # Display original image
cv2_imshow(downsampled_image_half)  # Display downsampled (1/2)
cv2_imshow(downsampled_image_quarter)  # Display downsampled (1/4)
```
