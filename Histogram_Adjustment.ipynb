```python
#sarinakasaiyan

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale

i = cv2.imread('/content/Histogram-Adjustment.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram of the original image

h1 = cv2.calcHist([i], [0], None, [256], [0, 256])

# Perform histogram equalization

g = cv2.equalizeHist(i)

# Calculate the histogram of the equalized image

h2 = cv2.calcHist([g], [0], None, [256], [0, 256])

# Display images and histograms

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.bar(range(256), h1.ravel())
plt.title('Original Histogram')
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(g, cv2.COLOR_BGR2RGB))
plt.title('Equalized Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.bar(range(256), h2.ravel())
plt.title('Equalized Histogram')


plt.show()
```
