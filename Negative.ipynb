```python
#sarinakasaiyan
import cv2
import matplotlib.pyplot as plt

# Read the image in grayscale

image_path = '/content/Negative.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image

plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axes
plt.show()

# Calculate the histogram of the original image

histogram_original = cv2.calcHist([original_image], [0], None, [256], [0, 256])

# Plot the histogram of the original image

plt.figure()
plt.bar(range(256), histogram_original.ravel(), color='black')
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 255])  # Set x-axis limits
plt.show()

# Create the negative image

negative_image = cv2.bitwise_not(original_image)
# Display the negative image
plt.imshow(negative_image, cmap='gray')
plt.title('Negative Image')
plt.axis('off')  # Hide axes
plt.show()

# Calculate the histogram for the negative image

histogram_negative = cv2.calcHist([negative_image], [0], None, [256], [0, 256])

# Plot the histogram for the negative image

plt.figure()
plt.bar(range(256), histogram_negative.ravel(), color='black')
plt.title('Histogram of Negative Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 255])  # Set x-axis limits
plt.show()
```
